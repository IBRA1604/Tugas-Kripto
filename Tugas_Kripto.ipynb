```python
def xor_encrypt(plain_text, key):
    encrypted = ""
    key_length = len(key)
    for i, char in enumerate(plain_text):
        encrypted += chr(ord(char) ^ ord(key[i % key_length]))
    return encrypted

# Main program
from google.colab import files

# Upload plain.txt file
uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, 'r') as fin:
    plain_text = fin.read()

key = input("Enter the encryption key: ")

encrypted_text = xor_encrypt(plain_text, key)

# Save encrypted text to cipher.txt
with open('cipher.txt', 'w') as fout:
    fout.write(encrypted_text)

print("Encryption complete. Encrypted text saved to 'cipher.txt'")

# Download the encrypted file
files.download('cipher.txt')
```

```python
def xor_decrypt(cipher_text, key):
    decrypted = ""
    key_length = len(key)
    for i, char in enumerate(cipher_text):
        decrypted += chr(ord(char) ^ ord(key[i % key_length]))
    return decrypted

# Main program
from google.colab import files

# Upload cipher.txt file
print("Please upload the cipher.txt file:")
uploaded = files.upload()
filename = next(iter(uploaded))

with open(filename, 'r') as fin:
    cipher_text = fin.read()

key = input("Enter the decryption key: ")

decrypted_text = xor_decrypt(cipher_text, key)

# Save decrypted text to plain2.txt
with open('plain2.txt', 'w') as fout:
    fout.write(decrypted_text)

print("Decryption complete. Decrypted text saved to 'plain2.txt'")

# Download the decrypted file
files.download('plain2.txt')
```
